## **Starting with structured Cleaning Tamplete which could be reused for future EDA.**

```python
# Header Cleaning or Columns check
print(df.columns.tolist())
print(f"No. of Columns present in Dataset is: " , len(df.columns.tolist()))

print("Fixing the Spaces withe the _ value in column names")
# df.columns = df.columns.str.lower() or .str.upper() as per requirement
df.columns = df.columns.str.replace(' ', '_')
print(df.columns.tolist())
```

```python
# DataType COnversion Here Spend in this case
df.Spend.head(5)
# df.Spend.sample(5)
# therefore we need to remove the $ sign to make same type of Spend Currrency
df['Spend'] = df['Spend'].astype(str).str.replace('$', '', regex=False)
df['Spend'] = pd.to_numeric(df['Spend'])
df.Spend.head(5)
# df.Spend.dtype
```

```python
# Categorical Analysis of the Column - Duplicate/Redundant Columns
print(df.Channel.unique())
print(df.Channel.value_counts())

# Some columns Like Facebook Instagram and Tiktok are repeated
```

```python
cleanup_map = {
    'TikTok' : 'Tik_Tok',
    'Facebook' : 'Facebok',
    'Instagram' : 'Insta_gram',
    'Gogle' : 'Google Ads',
    'E-mail' : 'Email',
    'N/A' : np.nan
}

df.Channel = df.Channel.replace(cleanup_map)
print(df.Channel.unique())
df.Channel
```

```python
# Handling Mixed Boolean Values
print(df.Active.unique())
```

```python
bool_map = {
    'Y': True,
    'Yes': True,
    '1': True,
    'True': True,
    'False': False,
    'No': False,
    '0': False
}

df['Active'] = df['Active'].replace(bool_map)
df['Active'] = df['Active'].astype(bool)
print(df.Active.unique())
```

# Handling Missing Values

## Inspect Missing Values and Impute Channel Column

Display the number of missing values for each column to understand the extent of the problem. df.isnull().sum()



```python
# claculating mode of channel column
channel_mode = df['Channel'].mode()[0]
# Mode imputation of Channel
df['Channel'] = df['Channel'].fillna(channel_mode)
print(f"Missing values in 'Channel' after imputation: {df['Channel'].isnull().sum()}")
```

## Impute 'Conversions' Column using the median




```python
# claculating the median of conversion column
conversions_median = df['Conversions'].median()
# Median Imputation
df['Conversions'] = df['Conversions'].fillna(conversions_median)
print(f"Missing values in 'Conversions' after imputation: {df['Conversions'].isnull().sum()}")
```

## Impute Sparse 'Clicks' Column with '0' value using Fillna



```python
df['Clicks'].fillna(0, inplace=True)
print(f"Missing values in 'Clicks' after imputation: {df['Clicks'].isnull().sum()}")
```

```python
df['Clicks'] = df['Clicks'].fillna(0)
print(f"Missing values in 'Clicks' after imputation: {df['Clicks'].isnull().sum()}")
```

```python
# Verifying All the Missing values has been filled
print("Missing values after all imputations:")
df.isnull().sum()
```

```python
# Date-time Parsing
print(df.Start_Date.head())
print(df.End_Date.head())

print(df.Start_Date.dtype)
print(df.End_Date.dtype)

# format : df['date'] = pd.to_datetime(df['date'], format='%d-%m-%Y')
df.Start_Date = pd.to_datetime(df.Start_Date, errors='coerce')
df.End_Date = pd.to_datetime(df.End_Date, dayfirst=True)

print(df.Start_Date.head())
print(df.End_Date.head())
```

```python
print(df.columns.tolist())
df = df.drop('Clicks', axis=1)
print("Columns after removing the redundant 'Clicks' column:")
print(df.columns.tolist())
```

```python
# Lgical integrity such as there is no clicks more than the no. of impressions
clicks_count = df.Clicks_.count()
impressions_count = df.Impressions.count()
if clicks_count <= impressions_count:
    print("There is no clicks more than the no. of impressions")
```

```python
# similarly we can check the logical integrity for the time traversal like if end_date < start_date
time_travel_mask = df.End_Date < df.Start_Date
if time_travel_mask.any():
    print("There is time travel in the dataset.")
    df.loc[time_travel_mask, 'End_Date'] = df.loc[time_travel_mask, 'Start_Date'] + pd.Timedelta(days=30)
    print("Time travel corrected.")
else:
    print("No time travel found in the dataset.")
```

```python
# Handling Outliers by IQR method in Spend column

Q1 = df.Spend.quantile(0.25)
Q3 = df.Spend.quantile(0.75)
IQR = Q3 - Q1
upper_fence = Q3 + 1.5 * IQR
lower_fence = Q1 - 1.5 * IQR

outlier_mask = (df.Spend > upper_fence) | (df.Spend < lower_fence)
print(df.loc[outlier_mask, 'Spend'])

# to fix this issue assign the outliers to upper or lower limit
df.loc[df['Spend'] > upper_fence, 'Spend'] = upper_fence
df.loc[df['Spend'] < lower_fence, 'Spend'] = lower_fence
```
